```python
# Задача 4. Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, 
# а среднее квадратическое отклонение равно 8 см. посчитайте, какова вероятность того, что случайным образом выбранный 
# взрослый человек имеет рост:
# 1. больше 182 см?
# 2. больше 190 см?
# 3. от 166 см до 190 см?
# 4. от 166 см до 182 см?
# 5. от 158 см до 190 см?
# 6. не выше 150 см или не ниже 190 см?
# 7. не выше 150 см или не ниже 198 см?
# 8. ниже 166 см?

from scipy import stats
P1 = stats.norm.cdf(182, loc=174, scale=8)
print(f'1. Вероятность того что взрослый человек имеет рост больше 182 см 1-P1 = {round((1-P1),3)}')
P2 = stats.norm.cdf(190, loc=174, scale=8)
print(f'2. Вероятность того что взрослый человек имеет рост больше 190 см 1-P2 = {round((1-P2),3)}')
P3 = stats.norm.cdf(166, loc=174, scale=8)
print(f'3. Вероятность того что взрослый человек имеет рост от 166 см до 190 см P2-P3 = {round((P2-P3),3)}')
print(f'4. Вероятность того что взрослый человек имеет рост от 166 см до 182 см P1-P3 = {round((P1-P3),3)}')
P4 = stats.norm.cdf(158, loc=174, scale=8)
print(f'5. Вероятность того что взрослый человек имеет рост от 158 см до 190 см P2-P4 = {round((P2-P4),3)}')
P5 = stats.norm.cdf(150, loc=174, scale=8)
print(f'6. Вероятность того что взрослый человек имеет рост не выше 150 см или не ниже 190 см 1-(P2-P5) = {round((1-(P2-P5)),3)}')
P6 = stats.norm.cdf(198, loc=174, scale=8)
print(f'7. Вероятность того что взрослый человек имеет рост не выше 150 см или не ниже 198 см 1-(P6-P4) = {round((1-(P6-P5)),4)}')
print(f'8. Вероятность того что взрослый человек имеет рост ниже 166 см 1-P3 = {round((1-P3),3)}')
print('Вероятность вычесленная с помощью билиотеки scipy совпадает рассчитаннами по правилу 3-х сигм')
```
